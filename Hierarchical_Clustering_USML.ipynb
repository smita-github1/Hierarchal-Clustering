```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
data=pd.read_csv("/content/diabetes.csv")
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
# to find out whether dataset has any null values
data.isnull().sum()
```

```python
# to extract outcome column
df=data.drop(['Outcome'],axis=1)
```

```python
import scipy.cluster.hierarchy as sch
plt.title("Diabetes Dendrogram")
dendrogram=sch.dendrogram(sch.linkage(df,method='ward'))
```

observation:  From above Dendrogram we can cluster patients into diabetics and non-diabetics.

 Orange color represents Diabetic patients


 Green color represents non-diabetics.

 Apart from this we can use dendrogram for determining the K- Value in K-Means
 algorithm.Draw a dendrogram and see how many clusters are created.

```python
# creating dendrogram on country dataset to indetify the number of clusters
country=pd.read_csv("/content/Country clusters.csv")
country.head()
```
